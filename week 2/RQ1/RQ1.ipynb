```python
import pandas as pd
df = pd.read_excel('labeling agreement.xlsx')
```

```python
df['is_match_approach_1']=df['category labled by Waleed']==df['category labled by Ameer']
```

```python
df_True=df[df['is_match_approach_1']==True]
```

```python
df_False=df[df['is_match_approach_1']==False]
```

```python
Waleed_list = df['category labled by Waleed'].tolist()
Waleed_list_l=[]
kappa_list_waleed=[]
for i in Waleed_list:
    if type(i) is str :
        kappa_list_waleed.append(99)
        l=i.split(',')
        li=[int(x) for x in l]
        for k in li :
            Waleed_list_l.append(k)
    else:
        Waleed_list_l.append(i)
        kappa_list_waleed.append(i)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# define a list of values
values = Waleed_list_l

# count the frequency of each value
value_counts = Counter(values)

# extract the values and their counts as separate lists
labels, counts = zip(*value_counts.items())

# create a bar plot
plt.bar(labels, counts)

# set the title and labels for the plot
plt.title('Distribution of labels used by Waleed')
plt.xlabel('Label')
plt.ylabel('Frequency')

# display the plot
plt.show()
```

## Qualitative analysis : Inter-rater agreement

```python
from sklearn.metrics import cohen_kappa_score

kappa_list_Ameer = df['category labled by Ameer'].tolist()

sample1 = kappa_list_Ameer
sample2 = kappa_list_waleed

# Calculate Cohen's kappa
kappa = cohen_kappa_score(sample1, sample2)

# Print the kappa score
print("Cohen's kappa: {:.2f}".format(kappa))
```
